```python
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'OpenAlex Publication Year' to numeric and handle any errors
df['publication_year'] = pd.to_numeric(df['OpenAlex Publication Year'], errors='coerce')

# Remove rows with missing publication years or database mentions, and filter starting from 1995
df_year_database = df[['publication_year', 'all_databases_mentioned']].dropna()
df_year_database = df_year_database[df_year_database['publication_year'] >= 1995]

# Create a dictionary to store database usage counts for each 5-year period
year_database_counts = defaultdict(lambda: defaultdict(int))

# Group years into 5-year periods and count databases
for index, row in df_year_database.iterrows():
    year = row['publication_year']
    period_start = (year // 5) * 5  # Group by 5-year periods
    if period_start >= 2025:  # Exclude the 2025 period
        continue
    for database in row['all_databases_mentioned'].split(','):
        database = database.strip()  # Ensure uniformity (lowercase)
        year_database_counts[period_start][database] += 1

# Aggregate the total usage of each database
database_totals = defaultdict(int)
for period in year_database_counts:
    for database, count in year_database_counts[period].items():
        database_totals[database] += count

# Get the top 20 databases overall
top_20_databases = sorted(database_totals, key=database_totals.get, reverse=True)[:20]

# Create subplots for each 5-year period, 3 graphs per row, sorted from recent to old
periods = sorted(year_database_counts.keys(), reverse=True)  # Sort from recent to old
num_periods = len(periods)
num_cols = 3
num_rows = (num_periods + num_cols - 1) // num_cols  # Calculate the number of rows needed

fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 6))
axs = axs.flatten()  # Flatten axes array for easier iteration

# Set a unified color palette for better consistency
colors = sns.color_palette("Blues", len(top_20_databases))

# Plot each period's data
for i, period_start in enumerate(periods):
    # Get the counts for the top 20 databases in the current period
    counts = [year_database_counts[period_start].get(database, 0) for database in top_20_databases]

    # Filter out databases with count 0
    filtered_databases_counts = [(top_20_databases[idx], counts[idx]) for idx in range(len(counts)) if counts[idx] > 0]
    if not filtered_databases_counts:  # Skip if no databases have non-zero counts
        continue

    sorted_databases, sorted_counts = zip(*sorted(filtered_databases_counts, key=lambda x: x[1], reverse=True))

    # Adjust figure size based on the number of databases
    bars = axs[i].barh(sorted_databases, sorted_counts, color=colors[:len(sorted_databases)])
    axs[i].set_xlabel('Number of Mentions', fontsize=12)
    axs[i].set_ylabel('Databases', fontsize=12)
    axs[i].set_title(f'Database Usage Trends ({period_start}-{period_start+4})', fontsize=14, fontweight='bold')
    axs[i].invert_yaxis()  # Invert y-axis to have the largest value at the top

    # Add value labels to each bar for better engagement
    for bar_idx, bar in enumerate(bars):
        axs[i].text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{int(bar.get_width())}', va='center', fontsize=10, color='black')
        # Highlight the top 3 databases
        if bar_idx < 3:
            bar.set_color('gold')
            bar.set_edgecolor('black')
            bar.set_linewidth(1.5)

# Remove any empty subplots
for j in range(i+1, len(axs)):
    fig.delaxes(axs[j])

# Add a main title to the entire figure to set the context for the LinkedIn post
fig.suptitle('Evolution of Research Database Usage in Bibliometric and Scientometric Studies (1995-2024)', fontsize=16, fontweight='bold', y=1.0)
plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()
```
